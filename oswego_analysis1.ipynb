```python
from datetime import datetime

def convert24(time):
    # Parse the time string into a datetime object
    t = datetime.strptime(time, '%I:%M %p')
    # Format the datetime object into a 24-hour time string
    return t.strftime('%H:%M')
```

```python
cdf['DATE OF ONSET'] = cdf['DATE OF ONSET'].apply(lambda x: x.replace('-Apr',''))
```

```python
cdf.head()
```

```python
cdf['TIME OF MEAL'] = cdf['TIME OF MEAL'].apply(lambda x: convert24(x))
```

```python
cdf.head()
```

```python
cdf['TIME OF ONSET'] = cdf['TIME OF ONSET'].apply(lambda x: convert24(x))
```

```python
cdf
```

```python
def convert_min_to_fraction(x):
    x= x.replace(':','.')
    x=x.replace('15','25')
    x=x.replace('30','50')
    x=x.replace('45','75')
    return x
```

```python
convert_min_to_fraction('19:30')
```

```python
cdf.head()
```

```python
cdf['TIME OF ONSET'] = cdf['TIME OF ONSET'].apply(lambda x: convert_min_to_fraction(x))
```

```python
cdf.head()
```

```python
cdf.loc[cdf['DATE OF ONSET']=='19','TIME OF ONSET'] =  cdf.loc[cdf['DATE OF ONSET']=='19','TIME OF ONSET'].astype('float') +12
```

```python
cdf['TIME OF ONSET'] = cdf['TIME OF ONSET'].astype('float')
```

```python
cdf['incubation'] = cdf['TIME OF ONSET'] - cdf['TIME OF MEAL']
```
