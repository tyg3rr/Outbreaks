```python
from datetime import datetime

def convert24(time):
    # Parse the time string into a datetime object
    t = datetime.strptime(time, '%I:%M %p')
    # Format the datetime object into a 24-hour time string
    return t.strftime('%H:%M')


def convert_min_to_fraction(x):
    x= x.replace(':','.')
    x=x.replace('15','25')
    x=x.replace('30','50')
    x=x.replace('45','75')
    return x
```

```python
main = df.copy()
```

```python
df = main.copy()
```

```python
df = df.loc[df['ILL']=='Y'][['TIME_OF_MEAL','ONSET_DATE','ONSET_TIME']]
```

```python
df = df.loc[df['TIME_OF_MEAL']!='unk']
```

```python
df['ONSET_TIME'] = df['ONSET_TIME'].apply(lambda x: x.replace('*',''))
```

```python
df['TIME_OF_MEAL'] = df['TIME_OF_MEAL'].apply(lambda x: x.replace('PM',' PM'))
```

```python
df
```

```python
for x in ['TIME_OF_MEAL','ONSET_TIME']:
    df[x] = df[x].apply(lambda y: convert24(y))
    
```

```python
for x in ['TIME_OF_MEAL','ONSET_TIME']:
    df[x] = df[x].apply(lambda y: convert_min_to_fraction(y))
```

```python
for x in ['TIME_OF_MEAL','ONSET_TIME']:
    df[x] = df[x].astype('float')
```
